```python
# Matplotlib(맷플롯립)은
# 파이썬에서 데이터를 시각화하여 그래프로 나타내는 데 가장 널리 사용되는 라이브러리입니다.
# 라인 플롯, 바 차트, 히스토그램, 산점도 등 거의 모든 종류의 그래프를 그릴 수 있으며,
# 데이터 분석 결과를 시각적으로 표현하고 인사이트를 얻는 데 필수적입니다.

# 기본 개념 및 설명
# 1. Matplotlib의 역할 🎨
#
# Matplotlib는 숫자 데이터를 그림으로 바꾸는 다리 역할을 합니다.
# 복잡한 숫자나 표만으로는 파악하기 어려운 데이터의 패턴, 추세, 관계 등을
# 그래프를 통해 한눈에 이해할 수 있게 도와줍니다.
#
# 2. 핵심 구성 요소
#
# Matplotlib 그래프는 보통 두 가지 주요 요소로 구성됩니다.
#
# Figure (피규어): 그래프가 그려지는 전체 캔버스나 종이입니다.
#하나의 Figure 안에는 여러 개의 그래프(Axes)가 포함될 수 있습니다.
#
# Axes (축): 실제 데이터가 그려지는 플롯 공간입니다.
# x축, y축, 눈금, 레이블 등 그래프의 모든 구성 요소를 포함합니다.
# '하나의 그래프'라고 생각하면 쉽습니다.

# 설치
# pip install matplotlib

# 사용법
# import matplotlib.pyplot as plt
```

```python
# 간단한 사용 예시: 꺾은선 그래프 그리기
# 가장 기본적인 꺾은선 그래프(Line Plot)를 그리는 과정은 다음과 같습니다.
#
# 1. 데이터 준비: x축과 y축에 해당하는 데이터를 준비합니다.
# (주로 리스트나 NumPy 배열 사용)
# 2. 플롯 생성: plt.plot() 함수를 사용해 데이터를 그래프로 그립니다.
# 3. 그래프 꾸미기 (선택 사항):
# plt.title(), plt.xlabel(), plt.ylabel() 등으로 제목과 축 이름을 추가합니다.
# 4. 그래프 보여주기: plt.show() 함수를 호출하여 화면에 그래프를 표시합니다.

import matplotlib.pyplot as plt

# 1. 데이터 준비 (x축: 월, y축: 월별 매출)
months = [1, 2, 3, 4, 5, 6]
sales = [150, 220, 280, 250, 310, 350]

# 2. 꺾은선 그래프 생성
plt.plot(months, sales)

# 3. 그래프 꾸미기
plt.title("Monthly Sales Trend") # 제목 추가
plt.xlabel("Month")              # x축 이름 추가
plt.ylabel("Sales (in millions)") # y축 이름 추가

# 4. 그래프를 화면에 표시
plt.show()
```
